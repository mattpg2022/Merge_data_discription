```python
## fianal version PG with highlight


from openpyxl.styles import PatternFill

def process_files(vendor_file, transaction_files):
    # Read the first file and create a dictionary of transaction codes and descriptions
    wb1 = openpyxl.load_workbook(vendor_file)
    sheet1 = wb1.active

    transaction_map = {}
    description_used = {}
    for row in range(5, sheet1.max_row + 1):
        transaction_code = sheet1.cell(row=row, column=5).value
        description = sheet1.cell(row=row, column=7).value
        amount = sheet1.cell(row=row, column=6).value

        if transaction_code:
            if transaction_code not in transaction_map:
                transaction_map[transaction_code] = []

            amount = float(amount) if amount is not None else None
            transaction_map[transaction_code].append({"description": description, "amount": amount, "row": row})
            description_used[row] = False

    # Read the transaction files, match the transaction codes and amounts, and add the descriptions in column L
    for transaction_file in transaction_files:
        wb2 = openpyxl.load_workbook(transaction_file)
        sheet2 = wb2.active

        for row in range(1, sheet2.max_row + 1):
            transaction_code = sheet2.cell(row=row, column=6).value
            transaction_amount = sheet2.cell(row=row, column=8).value
            positive_transaction_amount = sheet2.cell(row=row, column=9).value

            if transaction_code in transaction_map:
                combined_description = []
                for entry in transaction_map[transaction_code]:
                    if entry["amount"] is not None:
                        entry_amount = float(entry["amount"])
                        if (entry_amount < 0 and entry_amount == -positive_transaction_amount) or (entry_amount == transaction_amount):  # Check if the amounts match
                            if entry["description"] and any(c.isalnum() for c in entry["description"]):
                                combined_description.append(entry["description"])
                                description_used[entry["row"]] = True

                if combined_description:
                    new_description = ' | '.join(combined_description)
                    sheet2.cell(row=row, column=13).value = new_description
                    if len(combined_description) > 1:  # If more than one description was combined
                        yellow_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")
                        sheet2.cell(row=row, column=13).fill = yellow_fill

        # Save the changes to the transaction file
        wb2.save(transaction_file)

    # Highlight the unused description cells in the vendor file that contain text
    yellow_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")

    for row, used in description_used.items():
        description = sheet1.cell(row=row, column=7).value
        if not used and description and any(c.isalnum() for c in description):
            sheet1.cell(row=row, column=7).fill = yellow_fill

    # Save the changes to the vendor file
    wb1.save(vendor_file)

vendor_file = "18_PG_Vendor_Purchases_Details_April_24-30_2023.xlsx"
transaction_files = ["18_PG_Transactions_by_Account_April 24-30_2023.xlsx"]

process_files(vendor_file, transaction_files)
```
